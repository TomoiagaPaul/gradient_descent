```python
from linear_regression import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_all(X, Y, h=None, cost=None):
    import numpy as np
    import matplotlib.pyplot as plt

    X = np.array(X)
    
    plt.scatter(X, Y)
    if h is not None:
        plt.plot([X.min(), X.max()], h, c='red', linewidth=3, alpha=0.6)
    if cost is not None:
        plt.text(X.min()+0.3, h[0], 'Cost = ' + str(round(cost,2)), fontdict={'color':'red','size':'12', 'weight':'bold'})
    plt.title('Correlation of Housing Prices')
    plt.xlabel('Median income (US $10,000x)')
    plt.ylabel('Median house value (US $)')
```

```python
RAW = pd.read_csv('housing.csv')
RAW_c = RAW[RAW['median_house_value'] < 500000]
RAW_cs = RAW_c.sample(500)
data = RAW_cs

Y = data['median_house_value'] #replace with raw_cs
X = data['median_income']

plot_all(X, Y)
```

```python
X = [ [x] for x in X ]

linreg = LinearRegression(X, Y)
linreg.Train( learning_rate=0.0001 )

plot_all(X, Y,
         h=[linreg.Hypothesis( min(X) ) , linreg.Hypothesis( max(X) )],
         cost=linreg.Cost())
```
